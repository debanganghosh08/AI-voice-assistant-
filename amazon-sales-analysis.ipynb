# Sales per category

```python
category_sales = df.groupby('Category')['Total Sales'].sum().sort_values().to_frame()
category_sales
```

# Top 3 Products by Sales

```python
top_products= df.groupby('Product')['Total Sales'].sum().sort_values(ascending=False).to_frame().head(3)
top_products
```

# Average product price per category

```python
avg_price_category = df.groupby('Category')['Price'].mean().sort_values().to_frame()
avg_price_category
```

# Visualization

```python
plt.figure(figsize=(8,5))
sns.barplot(df, x='Category', y='Total Sales')
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=top_products, y='Product', x='Total Sales')
plt.title('Top 3 Products by Total Sales')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Total Sales'], bins=20, kde=True, color="purple")
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title('Correlation  of Telco dataset')
plt.show()
```
